STEP: 4 Exploratory Data Analysis

```python
gender_counts = data['Gender'].value_counts()
print(gender_counts)

gender_counts.plot(kind='bar',color= 'coral')

# Set labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution')

# Display the plot
plt.show()

print('The bar graph clearly indicates that the ratio of females to males is the same!!')
```

```python
females_yes = data[(data['Gender']== 'Female') & (data['Have you ever used the Christ bus facility?'] == 'Yes')]
males_yes = data[(data['Gender']== 'Male') & (data['Have you ever used the Christ bus facility?'] == 'Yes')]
print('Females who have used the Christ bus facility:',len(females_yes))
print('Males who have used the Christ bus facility:',len(males_yes))

counts = data.groupby(['Gender', 'Have you ever used the Christ bus facility?']).size()

# Create a bar plot
counts.plot(kind="bar")
plt.xlabel('Gender, Used bus or not?')
plt.ylabel('Count')
plt.title('Relationship between Gender and Bus Facility Usage')
plt.show()
```

```python
disp = data['Why do you prefer to go to Pune?'].value_counts()
print(disp)
print()
reasons_for_pune = data['Why do you prefer to go to Pune?']
most_com_reason = reasons_for_pune.mode().iloc[0]
min_count = data['Why do you prefer to go to Pune?'].value_counts().min()
min_value = disp[disp == min_count]

plt.hist(reasons_for_pune, bins=len(disp), edgecolor='k', alpha=0.7, color = 'green')
plt.xlabel('Reasons for Visiting Pune')
plt.ylabel('Frequency')
plt.title('Frequency Distribution of Reasons for Visiting Pune')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
print('The most common reason to visit Pune:', most_com_reason)
print()
print('The least commom reason for:',min_value)
```

```python
result = data[data['Do you use the 24/7 stay back facility of the Central Campus?'] == 'Yes'].groupby('Programme ')['Programme '].size()
print(result,'\n')
max_dept = result.idxmax()
max_stay = result.max()

result.plot(kind='bar', color='purple')

# Add labels and title
plt.xlabel('Departments')
plt.ylabel('No. of Students Staying back')
plt.title('Usage of 24/7 Stay Back Facility by Department')

# Display the plot
plt.xticks(rotation=90)  # Rotate x-axis labels for readability
plt.tight_layout()  # Adjust layout for better spacing
plt.show()

print('The dept from which the max students are staying back:', max_dept,'i.e',max_stay)
```

```python
females_yes = data[(data['Gender']== 'Female') & (data['For staycations in Pune from Saturday night, do you prefer a one-way bus service facility from the college?'] == 'Highly preferred')]
males_yes = data[(data['Gender']== 'Male') & (data['For staycations in Pune from Saturday night, do you prefer a one-way bus service facility from the college?'] == 'Highly preferred')]
print('Females who voted for staycations:',len(females_yes))
print('Males who voted for staycations',len(males_yes))

females_count = len(females_yes)
males_count = len(males_yes)

# Create a pie chart
labels = ['Females', 'Males']
sizes = [females_count, males_count]
colors = ['pink', 'lightblue']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%',startangle=140)
plt.title('Preference for One-Way Bus Service for Staycations')
plt.show()

print()
print('Clearly, boys voted more in comparison to girls!!')
```
